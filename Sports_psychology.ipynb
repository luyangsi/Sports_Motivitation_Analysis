```python
#Function definiation:Visualization
def plot_gender_stacked_percentage_bar(data, gender_col, reasons_cols):
    data[gender_col] = data[gender_col].replace({"": "Unknown"}).fillna("Unknown")
    for col in reasons_cols:
        data[col] = pd.to_numeric(data[col], errors='coerce').fillna(0)

    # groupdata by gender
    grouped = data.groupby(gender_col)[reasons_cols].sum()

    # calculate the percentage
    grouped_percentage = grouped.div(grouped.sum(axis=1), axis=0) * 100

    ax = grouped_percentage.plot(kind="bar", stacked=True, figsize=(10, 6), cmap="viridis")

    # Add percentage
    for bar_group in ax.containers:
        for bar in bar_group:
            height = bar.get_height()
            if height > 0:
                ax.text(
                    bar.get_x() + bar.get_width() / 2,
                    bar.get_y() + height / 2,
                    f"{height:.1f}%",
                    ha="center", va="center", fontsize=15, color="white"
                )
    plt.title("Visualization", fontsize=14)
    plt.ylabel("Percentage(%)", fontsize=12)
    plt.xlabel("Gender", fontsize=12)
    plt.legend(title="Reason", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()

    plt.show()
```

```python
#Function Definiation:Extra question cleaning
def clean_and_count_values(df, output_csv, columns_to_clean):
    df.columns = df.columns.str.strip()

    missing_columns = [col for col in columns_to_clean if col not in df.columns]
    if missing_columns:
        raise ValueError(f"These columns do not exist: {missing_columns}")

    # 4. NA value collection
    na_patterns = re.compile(r"^(NA|Na|N/A|N\?A|na)$", re.IGNORECASE)

    # 5. Replace all NA values with pd.NA
    for col in columns_to_clean:
        df[col] = df[col].replace(na_patterns, pd.NA)

    # 6. Delete rows with NA
    df_cleaned = df.dropna(subset=columns_to_clean)

    # 7. give teh value counts for remaining rows
    freq_result = {}
    for col in columns_to_clean:
        freq_result[col] = df_cleaned[col].value_counts()

    freq_df = pd.concat(freq_result, axis=1)

    # 8.Update it as a CSV file
    freq_df.to_csv(output_csv)

    return freq_df
```

```python
#Self physical and mental evalution profile

#df = pd.read_csv("Survey Response.csv")
df['mental_status'] = df['On a scale of 1- 5, how would you evaluate your mental health overall (if you feel uncomfortable answering you may skip)'].apply(
    lambda x: 'High' if x > 2 else 'Low'
)
df['physical_status'] = df['On a scale of 1- 5, how would you evaluate your physical fitness(if you feel uncomfortable answering you may skip)'].apply(
    lambda x: 'High' if x > 2 else 'Low'
)

#defining high and low threshold

def determine_health_group(row):
    if row['mental_status'] == 'High' and row['physical_status'] == 'High':
        return 'Both High'
    elif row['mental_status'] == 'Low' and row['physical_status'] == 'Low':
        return 'Both Low'
    elif row['mental_status'] == 'High':
        return 'High Mental'
    elif row['mental_status'] == 'Low':
        return 'Low Mental'
    elif row['physical_status'] == 'High':
        return 'High Physical'
    else:
        return 'Low Physical'

# dividing all response into 6 groups based on self rated mental and physical scores

df['health_group'] = df.apply(determine_health_group, axis=1)

print("Health group distribution:")
print(df['health_group'].value_counts())
```

```python
#SCS algorithm

def calculate_scs_score(df):
    scs_prefix = "self-compassion scale(SCS)"
    scs_cols = [col for col in df.columns if col.startswith(scs_prefix)]
    # identify columns that are SCS questions

    result_df = df.copy()

    scs_data = df[scs_cols].copy()
    for col in scs_cols:
        scs_data[col] = pd.to_numeric(scs_data[col].str[0], errors='coerce')

    subscales = {
        'Self-Kindness': [2, 6],
        'Self-Judgment': [11, 12],  # reverse scored
        'Common-Humanity': [5, 10],
        'Isolation': [4, 8],  # reverse scored
        'Mindfulness': [3, 7],
        'Over-identification': [1, 9]  # reverse scored
    }

    for subscale, items in subscales.items():
        cols = [scs_cols[i-1] for i in items]
        if subscale in ['Self-Judgment', 'Isolation', 'Over-identification']:
            result_df[f'SCS_{subscale}'] = 6 - scs_data[cols].mean(axis=1)
        else:
            result_df[f'SCS_{subscale}'] = scs_data[cols].mean(axis=1)

    subscale_cols = [f'SCS_{subscale}' for subscale in subscales.keys()]
    result_df['SCS_Total_Score'] = result_df[subscale_cols].mean(axis=1)

    result_df['SCS_Level'] = pd.cut(result_df['SCS_Total_Score'],
                                   bins=[0, 2.49, 3.5, 5.0], # threshold based on original author
                                   labels=['Low', 'Moderate', 'High'],
                                   include_lowest=True)

    return result_df
```

```python
def print_scs_statistics(df):
    print("\nSelf-Compassion Scale Statistics:")
    print("-" * 50)

# summary statistics
    print("\nOverall Score Statistics:")
    print(f"Mean Score: {df['SCS_Total_Score'].mean():.2f}")
    print(f"Median Score: {df['SCS_Total_Score'].median():.2f}")
    print(f"Standard Deviation: {df['SCS_Total_Score'].std():.2f}")
    print(f"Range: {df['SCS_Total_Score'].min():.2f} - {df['SCS_Total_Score'].max():.2f}")

# distribution of low, moderate and high
    print("\nDistribution across Levels:")
    level_dist = df['SCS_Level'].value_counts()
    for level, count in level_dist.items():
        percentage = (count/len(df)) * 100
        print(f"{level}: {count} participants ({percentage:.1f}%)")

    subscale_cols = [col for col in df.columns if col.startswith('SCS_')
                    and col not in ['SCS_Total_Score', 'SCS_Level', 'SCS_Percentile', 'SCS_Z_Score']]

# correlation of subscale scores and total score
    print("\nSubscale Correlations with Total Score:")
    correlations = df[subscale_cols].corrwith(df['SCS_Total_Score']).sort_values(ascending=False)
    for subscale, corr in correlations.items():
        print(f"{subscale.replace('SCS_', '')}: {corr:.3f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV files
stopping_reasons = pd.read_csv('Reasons Stopping from Participating in Sports.csv')
motivating_reasons = pd.read_csv('Reasons Motivating Participating in Sports.csv')
starting_reasons = pd.read_csv('reasons for starting sports.csv')

def analyze_categories(df, category_column, title):
    # Count categories
    category_counts = df[category_column].value_counts()

    # Calculate percentages
    category_percentages = (category_counts / len(df) * 100).round(1)

    # Create figure and axis
    plt.figure(figsize=(12, 6))

    # Create bar plot
    bars = plt.bar(range(len(category_counts)), category_counts)

    # Customize the plot
    plt.title(title, fontsize=14, pad=20)
    plt.xlabel('Categories', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(range(len(category_counts)), category_counts.index, rotation=45, ha='right')

    # Add value labels on top of each bar
    for i, (count, percentage) in enumerate(zip(category_counts, category_percentages)):
        plt.text(i, count, f'{count}\n({percentage}%)',
                ha='center', va='bottom')

    # Adjust layout to prevent label cutoff
    plt.tight_layout()

    # Print summary statistics
    print(f"\nSummary for {title}:")
    print("Category Counts:")
    for category, count in category_counts.items():
        percentage = category_percentages[category]
        print(f"{category}: {count} responses ({percentage}%)")
    print(f"Total responses: {len(df)}")

    plt.show()

# Analyze each dataset
analyze_categories(stopping_reasons, 'Categorization',
                  'Categories of Additional Reasons Stopping Sports Participation')

analyze_categories(motivating_reasons, 'Category',
                  'Categories of Additional Motivating Factors for Sports Participation')

analyze_categories(starting_reasons, 'Category',
                  'Categories of Reasons for Starting Sports')

# Create a combined visualization
plt.figure(figsize=(15, 8))

# Create three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

# Plot for stopping reasons
stopping_counts = stopping_reasons['Categorization'].value_counts()
ax1.pie(stopping_counts, labels=stopping_counts.index, autopct='%1.1f%%')
ax1.set_title('Stopping Reasons Distribution')

# Plot for motivating reasons
motivating_counts = motivating_reasons['Category'].value_counts()
ax2.pie(motivating_counts, labels=motivating_counts.index, autopct='%1.1f%%')
ax2.set_title('Motivating Factors Distribution')

# Plot for starting reasons
starting_counts = starting_reasons['Category'].value_counts()
ax3.pie(starting_counts, labels=starting_counts.index, autopct='%1.1f%%')
ax3.set_title('Starting Reasons Distribution')

plt.tight_layout()
plt.show()
```
